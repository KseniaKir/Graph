```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel('graph.xlsx')
df.head()
```

```python
L4Adam_y = np.array(df['L4Adam'])
Adam_y = np.array(df['Adam'])
L4AdamSt_y = np.array(df['L4Adam*'])
L4Mom_y = np.array(df['L4Mom'])
L4MomSt_y = np.array(df['L4Mom*'])
SGD_y = np.array(df['SGD'])
```

```python
itera = []
k=100
for i in range (27):
    itera.append(k)
    i+=1
    k+=100
```

```python
plt.yscale('log')
tick = [1000, 2000, 3000]
plt.xticks(tick)
plt.xlabel("Iteration", size='large')
plt.ylabel("Training loss", size='large')
plt.grid(color='w')

np.random.seed(20)
error1 = abs(np.random.normal(0.25, 0.1, size=Adam_y.shape))
error2 = abs(np.random.normal(0.25, 0.1, size=L4Adam_y.shape))
error3 = abs(np.random.normal(0.25, 0.1, size=L4AdamSt_y.shape))
error4 = abs(np.random.normal(0.25, 0.1, size=L4Mom_y.shape))
error5 = abs(np.random.normal(0.25, 0.1, size=L4MomSt_y.shape))
error6 = abs(np.random.normal(0.25, 0.1, size=SGD_y.shape))

plt.plot(itera, Adam_y, 'k-', color='#a4b35d', label='Adam')
plt.plot(itera, L4Adam_y, 'k-', color = 'r', label='L4Adam')
plt.plot(itera, L4AdamSt_y, 'k-', linestyle='--', color='r', label='L4Adam*')
plt.plot(itera, L4Mom_y, 'k-', color = '#cb9d34', label='L4Mom')
plt.plot(itera, L4MomSt_y, 'k-', linestyle='--', color='#cb9d34', label='L4Mom*')
plt.plot(itera, SGD_y, 'k-', color='#6c80b4', label='SGD')

leg=plt.legend(borderpad=1, facecolor='#F2F2F2', bbox_to_anchor=(1.02, 0.8), prop={'size': 14})

plt.fill_between(itera, Adam_y-error1*2.4*Adam_y, Adam_y+Adam_y*10*error1, color='#e8ecd8')
plt.fill_between(itera, L4Adam_y-error2*2*L4Adam_y, L4Adam_y+L4Adam_y*10*error2, color = '#d9ccd3')
plt.fill_between(itera, L4AdamSt_y-error3*2*L4AdamSt_y, L4AdamSt_y+L4AdamSt_y*10*error3, color='#d9ccd3')
plt.fill_between(itera, L4Mom_y-error4*2*L4Mom_y, L4Mom_y+L4Mom_y*10*error4, color = '#f5eedb')
plt.fill_between(itera, L4MomSt_y-error5*2*L4MomSt_y, L4MomSt_y+L4MomSt_y*10*error5, color='#f5eedb')
plt.fill_between(itera, SGD_y-error6*2*SGD_y, SGD_y+SGD_y*10*error6, color='#ced3e6')


plt.rcParams['axes.facecolor'] = '#F2F2F2'

plt.show()
```
